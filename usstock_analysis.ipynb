```python
import pandas as pd
df_stocks = pd.read_csv("/Volumes/ssd/AlgorithmicT/01_set_up_environment_and_data/us_stocks.csv")
```

```python
df_stocks.head(5)
```

```python
df_stocks['datetime'] =  pd.to_datetime(df_stocks['date'], format='%Y-%m-%d')
```

```python
df_stocks.head(5)
```

```python
df_stocks2000 = df_stocks[pd.DatetimeIndex(df_stocks['datetime']).year > 2000] 
```

```python
df_stocks2000.head(3)
```

```python
df_stocks2000.drop('date', axis=1, inplace=True)
```

```python
df_stocks2000.head(3)
```

```python
df_stocks2000.set_index(['ticker', 'datetime'])
```

```python
df_final = df_stocks2000.drop(['open','high','low','close','volume','ex-dividend','split_ratio'], axis=1)
```

```python
df_final.head(3)
```

```python
df_final.set_index(['ticker', 'datetime'])
```

```python
df_final.rename(columns = {'adj_open':'open', 'adj_high':'high','adj_low':'low','adj_close':'close','adj_volume':'volume'}, inplace = True)
```

```python
df_final.head(3)
```

```python
h5File = "us_stocks_hdf5"
df_final.to_hdf(h5File, "/Volumes/ssd/AlgorithmicT/01_set_up_environment_and_data/");
```
